```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/StudentsPerformance.csv')  # Replace with the correct path

# Combine the subject scores into one column for easier visualization
df['total_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Plot histogram of total scores
plt.figure(figsize=(10, 6))
plt.hist(df['total_score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Student Scores')
plt.xlabel('Total Scores')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Plot gender-wise performance comparison using a boxplot for math score
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math score', data=df, palette='Set2')
plt.title('Gender-wise Performance in Math')
plt.xlabel('Gender')
plt.ylabel('Math Scores')
plt.grid(True)
plt.show()

# Repeat for reading and writing score if needed
```

```python
# Plot scatter plot of parental level of education vs math score
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='math score', data=df, palette='Set3')
plt.title('Performance by Parental Level of Education (Math Score)')
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Scores')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Calculate average scores for each subject
subject_averages = df[['math score', 'reading score', 'writing score']].mean()

# Plot bar chart
plt.figure(figsize=(10, 6))
subject_averages.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Subject-wise Average Performance')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Define the grade function
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 75:
        return 'B'
    elif score >= 50:
        return 'C'
    else:
        return 'Fail'

# Calculate the total score and assign grades
df['total_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
df['grade'] = df['total_score'].apply(get_grade)

# Plot pie chart for grade distribution
grade_counts = df['grade'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], startangle=140)
plt.title('Distribution of Student Grades')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```
